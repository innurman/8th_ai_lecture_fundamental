## Multi Index

### stack

```python
tips.groupby('sex').mean()
```

`stack`은 컬럼을 인덱스로 만든다. 이때, 인덱스는 계층적인 구조를 가진 멀티인덱스가 된다.

```python
tips.groupby('sex').mean().stack() # SCI (Stack은 Column을 Index로 바꾼다.)
```

Series에서 괄호 한 개일 때는 행을 뽑는다.

```python
# 행 인덱스가 Male인 것을 뽑음.
# Male의 하위 인덱스인 total_tip, size는 남음.
tips.groupby('sex').mean().stack()['Male']
```

괄호 두 개일 때는 fancy indexing과 같은 개념이다.

```python
#  Male이 남아있다.
# fancy indexing하면 원래의 차원을 유지한다.
tips.groupby('sex').mean().stack()[['Male']]
```

### unstack

`unstack`은 인덱스를 컬럼으로 만든다. 

```python
tips.groupby('sex').mean()
```

인덱스였던 Male, Female이 컬럼의 하위 인덱스가 되었다.

```python
# index가 없어지므로 Series가 됨.
tips.groupby('sex').mean().unstack() # UIC : Unstack은 Index를 Column으로 바꾼다.
```

멀티인덱스인 상태에서 `unstack`하면 컬럼이 계층적인 구조를 가진 멀티인덱스가 된다.

```python
tips.groupby(['sex', 'day']).mean() 
```

sex 컬럼이 기존 컬럼의 하위 인덱스가 되었다.

```python
tips.groupby(['sex', 'day']).mean().unstack() 
```

멀티인덱스일 때 컬럼으로 바꿀 인덱스를 지정할 수 있다.

```python
tips.groupby(['sex', 'day']).mean().unstack(0) # default : level=-1
```

## 추천시스템 복습

```python
movies = pd.read_csv('ml-1m/movies.dat', 
                     sep='::', 
                     engine='python', # 원래 C 기반으로 데이터를 불러오는데, 빠르긴 하지만 유연하지 못하다. 
                                      # python 기반으로 불러오면 조금 느리지만 유연하다.
                     header=None,
                     names=['MovieID', 'Titles', 'Genres']
                    )
```

```python
users = pd.read_csv('ml-1m/users.dat', 
                    sep="::", 
                    engine="python", 
                    header=None,
                    names= ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'])
```

```python
ratings = pd.read_csv('ml-1m/ratings.dat', 
                    sep="::", 
                    engine="python", 
                    header=None,
                    names= ["UserID", "MovieID", "Rating", "Timestamp"])
```

```python
# 불러온 데이터를 하나로 합침
data = ratings.merge(users).merge(movies)
```

```python
# 원본을 유지하기 위해서 copy
recommendation_data = data[['UserID', 'MovieID', 'Rating']].copy()
```

```python
# UserID를 행, MovieID를 열, Rating을 값으로 하는 DataFrame 
recommendation_pivot = recommendation_data.pivot('UserID', 'MovieID', 'Rating')
```

```python
# 결측값(missing value)를 0으로 채움
recommendation_pivot.fillna(0, inplace=True)
```

```python
# 데이터가 많아서 상관관계를 구하는데 시간이 올래 걸리므로, 일부만 사용함
# corr은 열 기준으로 상관관계를 분석
# UserID가 열이 되어야 하므로 T(전치)
small_test = recommendation_pivot.T.iloc[:500,:500]
```

```python
# 유저들 간의 상관관계 분석
small_test_corr = small_test.corr()
```

```python
# 5번 유저와 상관관계가 높은 4명
# loc는 이름으로, iloc는 인덱스로 인덱싱
small_test_corr.loc[5].nlargest(4)
```

```python
# nlargest를 사용하지 않고 상관관계가 높은 상위 10명 뽑기 
small_test_corr.loc[5].sort_values(ascending=False)[:10]
```

```python
# 재할당하면 tab 자동완성을 사용할 수 있다. (편법)
x = recommendation_pivot.loc[5]
```

```python
x
```

```python
# Error
recommendation_pivot.T[x.where(x>0, False)]
```

pandas에서 `where`의 결과는 True, False가 아니라서 Boolean Indexing할 수 없다.

```python
# 조건이 참이면 값 반환, 거짓이면 False로 값을 바꾼다.
x.where(x>0, False).value_counts()
```

```python
def nearest_user(user_id, n):
    """user_id를 인자로 넣으면 상관관계가 높은 사람 n명을 알려주는 함수"""
    return small_test_corr.loc[user_id].nlargest(n+1)[1:]
```
