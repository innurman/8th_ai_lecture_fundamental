#### 구조 바꾸기

##### merge

`merge`는 이름이 같은 column을 기준으로 데이터를 합친다.

```python
data = ratings.merge(users).merge(movies)
```

```python
data
```

필요한 column만 추린다.

```python
# fancy indexing은 view 이므로 copy한다. (원본 반영 x)
recommendation_data = data[['UserID', 'MovieID', 'Rating']].copy()
```

```python
recommendation_data
```

##### join

`join`은 index를 기준으로 데이터를 합쳐준다.  
`merge`와 비슷하지만, 중복되는 column 이름이 있으면 이름 구분자를 정해줘야한다.(`rsuffix`, `lsuffix`)

```python
# on='UserID'는 ratings에서 UserID column을 기준으로 삼는 옵션이다.
# other에서는 UserID column을 인덱스로 삼았다.
# 결국, ratings와 users 모두 UserID을 기준으로 join한다.
ratings.join(other=users.set_index('UserID'), on="UserID")
```

```python
# users에 UserID column이 있는 상태로 join하면 UserID라는 이름의 column이 양쪽 모두에 있기 때문에 Error
# rsuffix 옵션으로 이름을 바꿔서 해결한다.
ratings.join(users, "UserID", rsuffix="_")
```

UserID를 인덱스로, MovieID를 column으로 평점 표를 만든다.

```python
# pivot_table과 다름
recommendation_pivot = recommendation_data.pivot(index="UserID", 
                                                 columns="MovieID", 
                                                 values="Rating")
```

```python
recommendation_pivot
```

### 분석

```python
# UserID가 2746인 사람이 본 영화 목록
# boolean indexing
recommendation_pivot.loc[2746][recommendation_pivot.loc[2746]>0]
```

```python
# UserID가 2746인 사람이 매긴 평점별 영화 개수 
# 5점 매긴 영화 11개, 4점 매긴 영화 28개, 3점 매긴 영화 18개
recommendation_pivot.loc[2746][recommendation_pivot.loc[2746]>0].value_counts()
```

```python
# 일반화 시킬 수 있다. -> 함수화
def movie_seen(user_id):
    return recommendation_pivot.loc[user_id][recommendation_pivot.loc[user_id]>0]
```

```python
# 5번 유저가 본 영화 및 평점
movie_seen(5)
```

#### corr

`corr`은 column 기준으로 상관관계를 분석해준다.

```python
# corr : correlationship(상관관계)
users[['UserID', 'Age']].corr()
```

UserID를 기준으로 상관관계를 구할 것이므로, column이 UserID여야한다.

```python
# T : 전치 행렬
small_test = recommendation_pivot.T.iloc[:500, :500] # 데이터가 커서 빠른 실습을 위해서 크기를 줄임 (실무에서는 모든 데이터 사용)
```

```python
small_test_corr = small_test.corr()
```

```python
small_test_corr
```

UserID가 5인 사람과 상관관계가 높은 User 상위 10명을 뽑을 수 있다.

```python
# sort_values는 값의 크기순으로 데이터 순서를 정렬한다.
# asceding=False는 내림차순
small_test_corr.loc[5].sort_values(ascending=False)[:10]
```

`nlargest`로 같은 결과를 낼 수 있다. 

```python
# 코드가 더 짧음
small_test_corr.loc[5].nlargest(10)
```

일반화 가능, 함수화 가능.

```python
def nearest_user(user_id, n):
    return small_test_corr.loc[user_id].nlargest(n+1)[1:] # 1번째 부터 인덱싱 : 자기자신(0번째)은 제외
```

만들어진 함수로 5번 유저와 상관관계가 높은 유저 5명을 찾을 수 있다.

```python
nearest_user(5, 5)
```

상관관계가 가장 높은 193번 유저와 5번유저가 본 영화를 비교한다.

```python
# 5번 유저가 본 영화
movie_seen(5).index
```

```python
# 193번 유저가 본 영화
movie_seen(5).index
```

```python
# 193번은 봤지만 5번 유저는 안 본 영화
unseen =set(movie_seen(193).index) - set(movie_seen(5).index)
```

```python
len(unseen)
```

193번은 봤지만 5번 유저가 안 본 영화를 추천해줄 수 있다.  
(취향이 비슷한 사람이 본 영화를 추천)

```python
# isin -> predicate -> boolean indexing
unseen_movie = recommendation_data[recommendation_data.MovieID.isin(unseen)]
```

```python
unseen_movie
```

193번 유저가 5점을 준 영화만 추천하면 더 만족도가 높을 것이다.  
(취향이 비슷한 사람이 평점을 잘 준 영화 추천)

```python
# 193번 유저가 5점 준 영화
unseen_movie[(unseen_movie.Rating == 5) & (unseen_movie.UserID == 193)]
```

`merge`로 영화 제목 확인할 수 있다.

```python
unseen_movie[(unseen_movie.Rating == 5) & (unseen_movie.UserID == 193)].merge(movies)
```

일반화할 수 있다.

```python
def unseen_movie(user_id, other_id):
    unseen = set(movie_seen(other_id).index) - set(movie_seen(user_id).index)
    unseen_movie = recommendation_data[recommendation_data['MovieID'].isin(unseen)]
    return unseen_movie[(unseen_movie.Rating == 5) & (unseen_movie.UserID == other_id)].merge(movies)
```

```python
unseen_movie(5, 193)
```

### 실습

Me: 7 / 유사한 2명 뽑아서 각각 평점 5점인 영화 추천

```python
def movie_seen(user_id):
    return recommendation_pivot.loc[user_id][recommendation_pivot.loc[user_id]>0]
```

```python
def nearest_user(user_id, n):
    return small_test_corr.loc[user_id].nlargest(n+1)[1:] # 1번째 부터 인덱싱 : 자기자신(0번째)은 제외
```

7번 유저와 유사도가 높은 2명을 찾는다.

```python
nearest_user(7, 2)
```

각각의 유저를 user0, user1로 해둔다.

```python
user0 = nearest_user(7,2).index[0]
user1 = nearest_user(7,2).index[1]
print(user0, user1)
```

user0이 본 영화, user1이 본 영화를 나열한다.

```python
movie_seen(user0).index
```

```python
movie_seen(user1).index
```

user0과 user1이 본 영화 중에서 겹치는 영화만 찾는다.

```python
# 집합에서 &는 교집합
intersect_movie = set(movie_seen(user0).index) & set(movie_seen(user1).index)
```

```python
intersect_movie
```

그 중에서7번 유저가 본 영화는 제외한다.

```python
unseen = intersect_movie - set(movie_seen(7).index)
```

```python
unseen
```

추천해줄 영화는 다음과 같다.

```python
# 288번, 251번 유저는 봤지만 5번 유저가 안본 영화 중에서
# 288번과 251번이 매긴 평점 데이터 중에서
# 평점 5점인 것만 추천

recommend = recommendation_data[recommendation_data['MovieID'].isin(unseen) \
                                # 288번, 251번 유저는 봤지만 5번 유저가 안본 영화
                                & (recommendation_data.UserID.isin([user0, user1])) \
                                # 288번과 251번이 매긴 평점
                                & (recommendation_data.Rating == 5)]          
                                # 평점 5점인 것만
```

```python
recommend
```

`merge` 해서 추천하는 영화의 제목과 장르를 알 수 있다.

```python
recommend.merge(movies)
```

일반화 시킬 수 있다.

```python
def recommend_movies(user_id):
    user0, user1 = nearest_user(user_id, 2).index
    intersect_movie = set(movie_seen(user0).index) & set(movie_seen(user1).index)
    unseen = intersect_movie - set(movie_seen(7).index)
    recommend = recommendation_data[recommendation_data['MovieID'].isin(unseen) \
                                & (recommendation_data.UserID.isin([user0, user1])) \
                                & (recommendation_data.Rating == 5)]          
    return recommend.merge(movies)
```

```python
recommend_movies(7)
```
